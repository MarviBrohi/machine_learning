```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(df["Salary"],df["YearsExperience"],test_size=0.2,random_state=42)

print(type(x_train))
print(type(x_train.values))
Test_array = np.random.randint(1,10,size=6).reshape(2,3)
Test_array
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
x_train = x_train.values.reshape(-1,1)
x_test = x_test.values.reshape(-1,1)
y_train = y_train.values.reshape(-1,1)
y_test = y_test.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
LinearR = LinearRegression()
LinearR.fit(x_train,y_train)
y_predict = LinearR.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error,root_mean_squared_error
r2 = r2_score(y_test, y_predict)
mse = mean_squared_error(y_test, y_predict)
MAE = mean_absolute_error(y_test,y_predict)
RMSE = root_mean_squared_error(y_test,y_predict)
print("R² Score:", r2)
print("Mean Squared Error:", mse)
print("Mean Absolute Error: ",MAE)
print("Root Mean Squared Error:",RMSE)
```

```python
m = LinearR.coef_
c = LinearR.intercept_
```

```python
plt.scatter(df["Salary"],df["YearsExperience"],color="yellow",label="Actual Data")
plt.plot(x_train,LinearR.predict(x_train),color="blue")
plt.show()
```
